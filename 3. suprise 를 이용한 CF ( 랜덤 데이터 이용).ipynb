<h2> surprise로 CF 구현 </h2>

```python
from surprise import SVD
from surprise import Dataset 
from surprise import accuracy 
from surprise.model_selection import train_test_split
from surprise import Reader
```

```python
ratings =  pd.read_csv('data/ratings_random.csv')
ratings.head()
```

<h3>surprise Reader class</h3>
raw data에서 surprise에서 사용하는 dataset로 사용하기 위한 로딩 규칙을 지정하기 위해 사용<br>
<h3>surprise Dataset class</h3>
surprise에서 dataset은 user item rating의 순서를 반드시 지켜야 한다

```python
# pandas DataFrame으로 데이터 로딩
reader =Reader(rating_scale = (1, 5))
data = Dataset.load_from_df(ratings[['store_id', 'influencer_id','rating']], reader)
```

```python
# file 에서 데이터 로딩
#reader = Reader(line_format='user item rating', sep=',', rating_scale=(1, 5), skip_lines=1)
#data=Dataset.load_from_file('data/ratings_random.csv',reader=reader)
```

```python
# 데이터를 train set과 test set으로 분리
# trainset -> 75% testset ->25%
trainset, testset = train_test_split(data, test_size=.25, random_state =0)
```

<h3> Model</h3>
algo = SVD() 모델 설정<br>
alog.fit()  학습<br>
predictions = algo.test(testset) 예측<br>
prediction = algo.predict(uid, iid) 단일 사용자 - 아이템 예측 <br>

```python
# 모델 설정 & 학습
# SVD : 행렬 분해 알고리즘 matrix factorization algorithm
algo = SVD()
algo.fit(trainset)
```

```python
predictions = algo.test(testset)
predictions[:5]
```

```python
[(pred.uid, pred.iid, pred.est) for pred in predictions[:3]]
```

```python
# 특정 사용자- 아이템에 대한 예측
# user id, item item은 문자열로 입력해야 한다
uid = str(20)
iid = str(10)
pred = algo.predict(uid, iid)
print(pred)
```

```python
# RMSE로 예측결과 대한 평가
accuracy.rmse(predictions)
```

<h3> hyper parameter 튜닝 </h3>

```python
#GridSearchCV 그리드를 사용한 복수 하이퍼 파라미터 최적화
from surprise.model_selection import GridSearchCV

#최적화할 파라미터들을 딕셔너리 형태로 지정
param_grid = { 'n_epochs':[20,40,60], 'n_factors':[50,100,150]}

# CV(cross-validation iterator)를 3개 폴드 세트로 지정, 평가는 rmse
gs = GridSearchCV(SVD, param_grid, measures = ['rmse'], cv =3)
gs.fit(data)

#최적 성능
print(gs.best_score['rmse'])
print(gs.best_params['rmse'])
```
