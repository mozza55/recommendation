<h2> 가게 상권 찾기 </h2>
<h3> 가게 좌표 값 저장 </h3>

```python
# area 좌표계 변환 값 저장
area = pd.read_csv('./data/area.csv')
area.columns=['area_level','area_level_name','area_code','area_name','x','y','sigungu_code','dong_code']
area['lat'] = np.nan #y
area['lon']= np.nan #x
area.head()
```

```python
#카카오api 호출
headers = {
    'Authorization': 'KEY VALUE',
}
for idx in area.index:
    params = (
        ('x', area.loc[idx]['x']),
        ('y', area.loc[idx]['y']),
        ('input_coord', 'WTM'),
        ('output_coord', 'WGS84'),
    )
    response = requests.get('https://dapi.kakao.com/v2/local/geo/transcoord.json', headers=headers, params=params)
    j = json.loads(response.text)
    WGS_y = j.get("documents")[0].get("y")
    WGS_x = j.get("documents")[0].get("x")
    area['lat'][idx]= WGS_y
    area['lon'][idx] = WGS_x

area.head()
```

```python
area.to_csv('./data/area.csv', index= False)
```

```python
def findArea(WGS_x,WGS_y):
    MIN = 999999
    index = -1
    for i in area.index:
        x = area['lon'][i]
        y = area['lat'][i]
        temp =math.pow(float(WGS_x)-x,2) +math.pow(float(WGS_y)-y,2)
        if temp<MIN:
            index = i
            MIN  = temp
    return index
index =findArea(WGS_x,WGS_y)
index
```

<h3> rating ( 유사도 측정 )</h3>

```python
def baseRecommend(id):
    ratings = []
    age1 =population.loc[stores.loc[id]['area_code'],['m','fm','age10','age20','age30','age40','age50','age60']].values
    #idx = 0
    for i in influencers.index:
        rating = 0
        # 카테고리 비교
        if int(stores.loc[id]['category_id']/100000) == int(influencers.loc[i]['category_id']/100000):
            #print(int(stores.loc[id]['category_id']/100000) ," ;",int(influencers.loc[i]['category_id']/100000))
            rating = rating + 3
        
        # 지역
        if int(stores.loc[id]['행정동코드']/100000000) == int(influencers.loc[i]['addr_code']/100000000):
            rating = rating +2
            if int(stores.loc[id]['행정동코드']/100000) == int(influencers.loc[i]['addr_code']/100000):
                rating = rating +1
        
        # 성별    
        gender = 1-math.pow(population.loc[stores.loc[id]['area_code']]['m'] - influencers.loc[i]['target_m'],2)/10000
        rating = rating + gender*2
        
        #연령
        age2 = influencers.loc[i,['target_m','target_w','target_10','target_20','target_30','target_40','target_50','target_60']].values
        #코사인유사도
        age = np.dot(age1,age2) /(np.linalg.norm(age1)*np.linalg.norm(age2))
        rating =rating +2*age
        ratings.append((i,rating))
        #df.loc[idx] = [i, rating]
        #idx = idx+1
    return ratings
```
