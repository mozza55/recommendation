```python
import pandas as pd
import requests
import json
import math
import numpy as np
```

```python
## 통합 인구 데이터 생성
area = pd.read_csv('./data/area.csv')

# 거주 & 직장 인구 데이터
area_work = pd.read_csv('./data/population_material/area_work_pop.csv') # 상권 직장인구
area_work = area_work.sort_values(['area_code'], ascending=True).reset_index(drop=True)
side_work = pd.read_csv('./data/population_material/areaside_work_pop.csv') # 상권 거주인구
side_work = side_work.sort_values(['area_code'], ascending=True).reset_index(drop=True)
area_live = pd.read_csv('./data/population_material/area_live_pop.csv') # 상권인근 직장인구
area_live = area_live.sort_values(['area_code'], ascending=True).reset_index(drop=True)
side_live = pd.read_csv("./data/population_material/areaside_live_pop.csv") # 상권인근 거주인구
side_live= side_live.sort_values(['area_code'], ascending=True).reset_index(drop=True)
```

```python
population = pd.DataFrame(columns = ['area_code', 'area_name','total','m','fm','age10','age20','age30','age40','age50','age60'])
population['area_code'] = area['area_code']
population['area_name'] = area['area_name']
population = population.sort_values(['area_code'], ascending=True).reset_index(drop=True)
population = population.fillna(0)
population['idx'] = population['area_code']
population=population.set_index('idx')
population
```

```python
area_work = area_work.set_index('area_code')
area_live = area_live.set_index('area_code')
side_work = side_work.set_index('area_code')
side_live = side_live.set_index('area_code')
```

```python
# 인구 수 합계
def populationSum(df):
    for i in df.index:
        population['total'][i] = population.loc[i]['total'] +df.loc[i]['total']
        population['m'][i] = population.loc[i]['m'] + df.loc[i]['m']
        population['fm'][i] = population.loc[i]['fm'] + df.loc[i]['fm']
        population['age10'][i] = population.loc[i]['age10'] + df.loc[i]['age10']
        population['age20'][i] = population.loc[i]['age20'] + df.loc[i]['age20']
        population['age30'][i] = population.loc[i]['age30'] + df.loc[i]['age30']
        population['age40'][i] = population.loc[i]['age40'] + df.loc[i]['age40']
        population['age50'][i] = population.loc[i]['age50'] + df.loc[i]['age50']
        population['age60'][i] = population.loc[i]['age60'] + df.loc[i]['age60']
```

```python
populationSum(area_live)
populationSum(area_work)
populationSum(side_live)
populationSum(side_work)
```

```python
# 결측 값 
population[population['total']==0]
```

```python
# 인구 종합 데이터
staying = pd.read_csv('./data/population.csv') #상주인구
floating = pd.read_csv('./data/population_floating.csv') #유동인구
staying=staying.set_index('area_code')
floating=floating.set_index('area_code')
floating
```

```python
# 인구 비율데이터 생성
# 상주 인구가 5000이하면 상주인구와 유동인구 비교 후 비율 계산
population = pd.DataFrame(columns = ['area_code', 'area_name','total','m','fm','age10','age20','age30','age40','age50','age60'])
population['area_code'] = area['area_code']
population['area_name'] = area['area_name']
population = population.sort_values(['area_code'], ascending=True).reset_index(drop=True)
population = population.fillna(0)
population['idx'] = population['area_code']
population=population.set_index('idx')
```

```python
#값대신 비율로 환산해서 넣어줌
def calRate(df):
    population['total'][i] = df.loc[i]['total']
    population['m'][i] = round(df.loc[i]['m']/df.loc[i]['total']*100,0)
    population['fm'][i] = round(df.loc[i]['fm']/df.loc[i]['total']*100,0)
    population['age10'][i] = round(df.loc[i]['age20']/df.loc[i]['total']*100,0)
    population['age20'][i] = round(df.loc[i]['age30']/df.loc[i]['total']*100,0)
    population['age30'][i] = round(df.loc[i]['age40']/df.loc[i]['total']*100,0)
    population['age40'][i] = round(df.loc[i]['age50']/df.loc[i]['total']*100,0)
    population['age50'][i] = round(df.loc[i]['age60']/df.loc[i]['total']*100,0)

for i in population.index:
    if i in staying.index:
        if staying.loc[i]['total'] <5000 and (i in floating.index) and staying.loc[i]['total'] < floating.loc[i]['total']:
            calRate(floating)
            continue
        calRate(staying)
    else:
        calRate(floating)
```

```python
population.to_csv('./data/population_rating.csv',index=False)
population
```
